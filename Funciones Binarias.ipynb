# Funciones Universales Binarias- Morales Martínez Adriel Josué1951

```python
import numpy as np

arr = np.array([5,36,17,18,9])

arr_2 = np.array([8,24,17,19,9])
```

```python
np.add(arr,arr_2)
```

```python
np.subtract(arr,arr_2)
```

```python
np.multiply(arr,arr_2)
```

```python
np.divide(arr,arr_2)
```

```python
np.array_equal(arr,arr_2)
```

```python
np.fmin(arr,arr_2)
```

```python
np.fmax(arr,arr_2)
```
